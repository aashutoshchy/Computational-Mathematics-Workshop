```python
# Import necessary modules

# Using math module for pi and sin function


import matplotlib.pyplot as plt

import math


# Step 1: Generate equidistant x-coordinates between the given interval

# Lists for coordinates

x = []

y = []


# Define start, end, and step size for x increment

start = -4 * math.pi

end = 4 * math.pi

step = 0.1 # Controls smoothness of the graph


i = start

while i <= end:

  x.append(i) # Store x-coordinate

# Step 2: Calculate the y-coordinate using sin function for each x

  y.append(math.sin(i))

  i += step


# Step 3: Plot the graph using calculated coordinates

# Label the graph properly

plt.plot(x, y, color="blue", label="sin function")

plt.xlabel("X")

plt.ylabel("Y")

plt.title("y = sin(x) ,-4π<= x <= 4π")

plt.grid(True)

plt.legend()

plt.show()
```

```python
# Task : Plot a circle with center at (a,b) and radius =r

# for a=b=2 and r = 4:

# Circle equation: x^2 + y^2 = r^2

# Paramateric equation:

# x =r*cos(theta) + a , and y = r*sin(theta) + b

x =[] # List to store x values

y= [] # List to store y values

r = 4

center = [2,2]

# paramater theta start value

theta_start = 0.0

# parameter Theta end value (bit more than 2pi)

theta_end = 2*math.pi

# step sizes to sample the paramater Theta

step_size= 0.125

#math.ceil() → round up to nearest integer

# Number of samples

steps = math.ceil((theta_end-theta_start)/step_size)

# initialize the theta

# variable to increment Theta by each step

theta = theta_start

# for loop to calculate x and y value at each steps

for i in range(0,steps):

  x.append(r*math.cos(theta)+center[0])

  y.append(r*math.sin(theta)+center[1])

theta += step_size

# create figure in new window

plt.figure()

# plot a dot a center

plt.plot(center[0],center[1],'.')

plt.plot(x,y,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

#plt.gca() → Get the current figure (axes object) for handling

ax =plt.gca()

#ensures 1 unit in x = 1 unit in y

#Without this ,figure is stretched(ircle could appear as an ellipse)

ax.set_aspect('equal',adjustable='box')

plt.show()
```

```python
# Exercise - 10

import matplotlib.pyplot as plt
import math

x =[] # List to store x values

y= [] # List to store y values

r = 4

center = [2, 2]

# Initializing the values
theta_start = 0.0
theta_end = 2 * math.pi
theta_step = 0.0125
theta = theta_start

# Figuring out how many steps it takes to complete
steps = math.ceil((theta_end - theta_start) / theta_step)

# Looping through the steps
for i in range(0,steps):
  x.append(r * math.cos(theta) + center[0])
  y.append(r * math.sin(theta) + center[1])
  theta += steps

plt.figure()
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.grid(True)
plt.plot(x, y, '-')
plt.show()
```

```python
# Exercise - 11
import matplotlib.pyplot as plt
import math
def Display_circle():
  x =[]

  y= []

  r = 4

  center = [2, 2]

  theta_start = 0.0
  theta_end = 2 * math.pi
  theta_step = 0.0125
  theta = theta_start

  steps = math.ceil((theta_end - theta_start) / theta_step)

  for i in range(0,steps):
    x.append(r * math.cos(theta) + center[0])
    y.append(r * math.sin(theta) + center[1])
    theta += steps
  return x, y


x = []

y= []

# Ellipse parameters

a=3 # semi-minor axis

b=5 # semi-major axis

center =[2,2] # center (p,q)


# Parameter theta

theta_start = 0.0

theta_end = 3*math.pi # make more than 2*pi

step_size = 0.125


# Number of samples

steps = math.ceil((theta_end -theta_start) / step_size)


# initializing theta

theta= theta_start


# calculate x and y cooradinates

for i in range(steps):

  x.append(a*math.cos(theta)+center[0]) # x= a*cos(θ) +p

  y.append(b*math.sin(theta)+center[1]) # y=b*sin(θ) +q

  theta += step_size # incremet the theta by step_size


# Plotting

plt.figure()

# plot the center point

plt.plot(center[0],center[1],'.')

#Plot the ellipse

plt.plot(x,y,'-')

x1, y1 = Display_circle()

plt.plot(x1,y1,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

# Equal aspect ration to avoid stretching

ax= plt.gca()

ax.set_aspect('equal',adjustable='box')

plt.show()
```
