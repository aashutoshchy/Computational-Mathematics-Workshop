Example 1: The Python math module contains trig functions but they only work in radians.
Write a function called deg_sin that prompts the user for an input in degree, converts it into
radian and calculates the sin value of the given input.

```python
# Example - 1
 # Approach - 1
## -- these are the components we need from the math library
from math import pi, sin
## -- a function that prompts user for an input converts it in radian and
# calculates sin value
def deg_sin():
  x = float(input("Please input a number "))
  answer = sin(pi*x/180)
  print("result = {}".format(answer))
deg_sin() ## -- call the function
```

```python
# Exercise - 5
def sumOfDivisors(n):
    total = 0

    for i in range(1, n):
        if n % i == 0:
            total += i

    return total
n = 7 # Going from 7 because we need to find the perfect number bigger than 6
while True:
  if sumOfDivisors(n) == n: # Perfect number condition
    print("The next perfect number is: ", n)
    break
  n += 1
```

```python
# Exercise - 6
marks = [20, 30, 35, 55, 70, 80, 90]

# Finding Mean
total = 0
for m in marks:
    total += m
mean = total / len(marks)

# Finding Median
n = len(marks)
if n % 2 == 1:
    median = marks[n // 2]
else:
    median = (marks[n // 2 - 1] + marks[n // 2]) / 2

print("Mean: ", mean)
print("Median: ", median)
```

```python
# Exercise - 7
def convert_to_binaryNum(n):
  binary = ""
  while n / 2 == 0:
    binary += str(n % 2)
    n = n // 2
  return binary[::-1] # String Reverse
print(convert_to_binaryNum(19))
```

```python
# Exercise - 8
def prodDigits(n):
  prod = 1 # Initialising to 1 because we have to multiply
  while n > 0:
    prod *= n % 10 # Multiplying with the remainder
    n = n // 10
  return prod
print(prodDigits(27))
```

```python
# Exercise - 9
def multiply(*args): # *args gets multiple arguments
  prod = 1 # Initialising to 1 because we have to multiply
  for i in args:
    prod *= i # Multiplying every argument
  return prod
print(multiply(1,2,3,4,5))
print(multiply(12, 13, 14))
```

```python
# Example - 3
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}

# a. Finding the length of set1
print("a. Length of set1:", len(set1))

# b. Finding the sum of elements in set 2
total = 0
for i in set2:
    total += i
print("b. Sum of set2:", total)

# c. Union
print("c. Union:", set1 | set2)

# d. Intersection
print("d. Intersection:", set1 & set2)

# e. set1 - set2
print("e. set1 - set2:", set1 - set2)

# f. set2 - set1
print("f. set2 - set1:", set2 - set1)

# g. Check subset
print("g. set3 subset of set1:", set3.issubset(set1))
print("   set3 subset of set2:", set3.issubset(set2))

# h. Complement of set1 with U = {1,...,50}
U = set()
for i in range(1, 51):
    U.add(i)

complement = U - set1
print("h. Complement of set1:", complement)

# i. Check if 15 is in set1
print("i. Is 15 in set1?", 15 in set1)

# j. Add 60 to set1
set1.add(60)
print("j. set1 after adding 60:", set1)

# k. Remove 60 from set2
set2.remove(60)
print("k. set2 after removing 60:", set2)

# l. Remove all elements from set3
set3.clear()
print("l. set3 after removing all elements:", set3)

# m. Find maximum and minimum in set2
max_val = max(set2)
min_val = min(set2)

print("m. Maximum in set2:", max_val)
print("   Minimum in set2:", min_val)
```
